```python
import numpy as np
import pandas as pd
```

```python
dfx = pd.read_csv('Diabetes_XTrain.csv')
dfy = pd.read_csv('Diabetes_YTrain.csv')
test = pd.read_csv('Diabetes_Xtest.csv')
```

```python
dfx.head()
```

```python
dfy.head()
```

```python
X = dfx.values
Y = dfy.values
T = test.values
```

```python
#KNN
def dist(x1,x2):
    return np.sqrt(sum((x1-x2)**2))
                   
def knn(X,Y,queryPoint,k=17):
    vals = []
    m = X.shape[0]
    for i in range(m):
        d = dist(queryPoint,X[i])
        vals.append((d,Y[i]))
    vals = sorted(vals)
    vals = vals[:k]
    
    vals = np.array(vals)
    new_vals = np.unique(vals[:,1],return_counts = True)
    
    index = new_vals[1].argmax()
    pred = new_vals[0][index]
    return pred
```

```python
diabetes_pred = []
for i in range(T.shape[0]):
    x = knn(X,Y,T[i])
    diabetes_pred.append(x[0])
```

```python
diabetes_pred
```

```python
op = pd.DataFrame({'Outcome':np.array(diabetes_pred)})
op.to_csv('Diabetes_Output.csv',index = False)
```
